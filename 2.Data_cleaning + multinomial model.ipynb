```python
import pandas as pd 

spamCollection = pd.read_csv("Assets/SMSSpamCollection", header = None, sep = "\t", names = ["Label","SMS"])
```

```python
randomizedSpamCollection = spamCollection.sample(frac = 1, random_state= 1)

# a train - test ratio of 80-20 will be used 


training_ratio = round(len(randomizedSpamCollection) * 0.8)

training_set = randomizedSpamCollection[:training_ratio].reset_index( drop = True )
test_set = randomizedSpamCollection[training_ratio:].reset_index( drop = True )
```

<p><b>List of things to do:</b></p>
<ul>
    <li>Tokenization</li>
    <li>Lowercasing</li>
    <li>Removing nonalphabetic</li>
</ul>
<p>I'm skipping stemming here because I dont want to confuse myself first lol. Leplace smoothing parameter will be set to 1</p>

```python
#clean here , training set 

def remove_numeric(corpus):

    for index, sentence in enumerate(corpus):
        currentSentence = sentence 
        currentSentenceList = currentSentence.split()
        newSentence = filter(lambda x: x.isalpha(), currentSentenceList)
        joinedNewSentence = " ".join(newSentence)
        corpus[index] = joinedNewSentence
```

```python
def lower_case(corpus):

    for index, sentence in enumerate(corpus):
        currentSentence = sentence 
        currentSentenceList = currentSentence.split()
        
        for index2,word in enumerate(currentSentenceList):
            currentSentenceList[index2] = word.lower()
            
        newSentence = " ".join(currentSentenceList)
        corpus[index] = newSentence
```

<p>Separating Spam and not spam from training set</p>

```python
trainingMessagesSpam = training_set[training_set["Label"] == "spam"].reset_index(drop = True)

trainingMessagesNotSpam = training_set[training_set["Label"] == "ham"].reset_index(drop = True)
```

<p><b>Not spam list has in total, 4815 words</b></p>
<p>We can use the log function to calculate the probabilities, and omit out the denominator because denominator is repeated and the same for every single calculation</p>

```python
ProbabilityOfNotSpam = 0.86541
ProbabilityOfSpam = 0.13459
```

<p>Now we can start testing our model! L O L</p>


```python
import math


def classifyMessage(message, result, result2, spam, notSpam):



    #determinine probability of spam first 
    spamProb = 1
    notSpamProb = 1
    messageList = message.split()
    remove_numeric(messageList)
    lower_case(messageList)
    for message in messageList:
        if message not in result.columns:
            spamProb *= 1/1394
        else:
            probability = result[message]["Total_Frequency"] / 1392
            spamProb *= probability
        
        if message not in result2.columns:
            notSpamProb *= 1/4817
        else:
            probability = result2[message]["Total_Frequency"] / 4815
            notSpamProb *= probability


    spamProb *= spam
    notSpamProb *= notSpam
    if spamProb > notSpamProb:
        return "spam"
    else:
        return "ham"

    
```

```python
#Calculating Accuracy 

test_set["result"] = test_set["SMS"].apply(classifyMessage, args = (result, result2, ProbabilityOfSpam, ProbabilityOfNotSpam))
test_set.head()
```

```python
correct = 0

for index, row in test_set.iterrows():

    correctResult = row[0]
    predicted = row[2]
    if correctResult == predicted:
        correct+=1


    
```
